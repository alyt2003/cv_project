```python
import cv2
import numpy as np
```

```python
def sobel_x(img, sigma=1.0,ksize=3):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    
   
  
    
    blurred = cv2.GaussianBlur(gray, (ksize, ksize), sigma)
    
   
    img_x = cv2.Sobel(blurred, cv2.CV_64F, 1, 0, ksize)
    
    
    output = cv2.convertScaleAbs(img_x)
    return output

def sobel_y(img, sigma=1.0,ksize=3):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    
    
    blurred = cv2.GaussianBlur(gray, (ksize, ksize), sigma)
    
   
    img_y = cv2.Sobel(blurred, cv2.CV_64F, 0, 1, ksize)
    
    
    output = cv2.convertScaleAbs(img_y)
    return output
def sobel_magnitude(img, sigma=1.0,ksize=3):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blurred = cv2.GaussianBlur(gray, (ksize, ksize), sigma)
    sobelx = cv2.Sobel(blurred, cv2.CV_64F, 1, 0, ksize)
    sobely = cv2.Sobel(blurred, cv2.CV_64F, 0, 1, ksize)
    magnitude = np.sqrt(sobelx**2 + sobely**2)
    mag = cv2.convertScaleAbs(magnitude)
    output = cv2.convertScaleAbs(mag)
    return output
```

```python


cap = cv2.VideoCapture(0)
mode='o'
sigma=1.0
while True:
    
    ret, frame = cap.read()
    

    if not ret:
        break
    ksize = int(6 * sigma + 1)  
    if ksize % 2 == 0:
        ksize += 1
  
    if mode == 'o':
        output=frame
    elif mode =='x':
        output=sobel_x(frame,sigma,ksize)
    elif mode=='y':
        output=sobel_y(frame,sigma,ksize)
    elif mode=='m':
        output=sobel_magnitude(frame,sigma,ksize)
    elif mode=='l':
        log = cv2.Laplacian(frame, cv2.CV_64F)
        output = cv2.convertScaleAbs(log)
    elif mode =='s':
        output=sobel_magnitude(frame,sigma,ksize)
        _, output = cv2.threshold(output, 80, 255, cv2.THRESH_BINARY)
    
    key = cv2.waitKey(1) & 0xFF
    if key == ord('q'):
        break
    elif key== ord('o'):
        mode='o'
    elif key ==ord('x'):
       mode='x'
    elif key ==ord('y'):
       mode='y'
    elif key ==ord('m'):
       mode='m'
    elif key ==ord('l'):
        mode='l'
    elif key ==ord('s'):
        mode='s'
    elif key== ord('+'):
        sigma += 0.5
    elif key == ord('-'):
        sigma =max(0.5,sigma-0.5)
    
    cv2.imshow("", output)



cap.release()
cv2.destroyAllWindows()
```
